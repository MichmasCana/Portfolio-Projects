```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://bulbapedia.bulbagarden.net/wiki/Ability'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find_all('table')[9]
```

```python
column_label = table.find_all('th')
print(column_label)
```

```python
column = [label.text.strip() for label in column_label]
print(column)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = column)
```

```python
df
```

```python
all_raw_row = table.find_all('tr')
print(all_raw_row)
```

```python
for data in all_raw_row[1:]:
    row310 = data.find_all('td')
    row = [data.text.strip() for data in row310]
    # print(row)

    length = len(df)
    df.loc[length] = row
```
